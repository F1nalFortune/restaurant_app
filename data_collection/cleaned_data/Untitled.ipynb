```python
import pandas as pd
import ast
import numpy as np
from tqdm.notebook import tnrange, tqdm
tqdm.pandas()



addresses = pd.read_csv("./addresses.csv", index_col=0)
ct_restaurants = pd.read_csv("./ct_restaurants.csv", index_col=0)
discover_ct = pd.read_csv("../discover_ct.csv", index_col=0)

restaurants = pd.read_csv("../restaurants.csv", index_col=0)
second_thread = pd.read_csv("../second_thread.csv", index_col=0)
state_data = pd.read_csv("../state_data.csv", index_col=0)
```

```python
def returnList(row):
  if type(row)==str:
    try:
      result = ast.literal_eval(row)
    except:
      result = row
    return result
  else:
    return np.nan
  
#fix data lists
data_lists = ['attributes', 'hours', 'pictures']
pbar = tqdm(data_lists)
for data in pbar:
    pbar.set_description(f"Processing {data}")
    ct_restaurants[data] = ct_restaurants[data].apply(returnList)
```
